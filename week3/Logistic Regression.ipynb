# Let's do the math that turns a 3rd class passenger's logit ration of dying into a probability

```python
import math
```

```python
logit_odds= -1.44677627 + 3 * 0.8501017
```

```python
odds = math.exp(logit_odds)
odds
```

```python
probability = odds / ( 1 + odds )
probability
```
