```python
def bed_colours(beds):
    if beds == 1: return 'red'
    if beds == 2: return 'green'
    if beds == 3: return 'blue'
    return 'pink'
    
```

```python
import matplotlib.pyplot
sacramento.plot.scatter('longitude', 'latitude', 
                        c=sacramento.beds.map(bed_colours))
```

```python
import matplotlib.pyplot
sacramento.plot.scatter('longitude', 'latitude', c=sacramento.price, colormap=matplotlib.pyplot.cm.inferno)
                        
```

```python
big_houses = sacramento[sacramento.beds > 4]
```

```python
big_houses.plot.scatter('latitude', 'longitude', 
                        c=big_houses.price, 
                        colormap=matplotlib.pyplot.cm.rainbow, 
                        s=big_houses.beds*50)
```

```python
sacramento.type.unique()
```

```python
sacramento['sale_day_of_year'] = sacramento.clean_date.dt.dayofyear
residential = sacramento[sacramento.type == 'Residential']
condo = sacramento[sacramento.type == 'Condo']
```

```python
(fig, ax) = matplotlib.pyplot.subplots()
ax.plot(residential.groupby('sale_day_of_year').price.max(), c='green', label='Residential')
ax.plot(condo.groupby('sale_day_of_year').price.max(), c='magenta', label='Condo', marker='^')
ax.set_xlabel('Day of year')
ax.set_ylabel('Price in ($)')
ax.legend(loc='right')
```
