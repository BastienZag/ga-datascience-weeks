```python
import pandas
import sklearn.datasets
```

```python
flowers = sklearn.datasets.load_iris()
flowers['data']
```

```python
import sklearn.cluster
```

```python
clusterer = sklearn.cluster.KMeans(n_clusters=3)
```

```python
cluster_suggestions = clusterer.fit_predict(flowers['data'])
```

```python
cluster_suggestions
```

```python
import sklearn.metrics
```

```python
sklearn.metrics.silhouette_score(flowers['data'], cluster_suggestions)
```

```python
clusterer10 = sklearn.cluster.KMeans(n_clusters=10)
cluster_suggestions10 = clusterer10.fit_predict(flowers['data'])
sklearn.metrics.silhouette_score(flowers['data'], cluster_suggestions10)
```

```python
scores = []
for i in range(2, 20):
    clustereri = sklearn.cluster.KMeans(n_clusters=i)
    cluster_suggestionsi = clustereri.fit_predict(flowers['data'])
    scores.append(sklearn.metrics.silhouette_score(flowers['data'], cluster_suggestionsi))
    
scores
```

```python
%matplotlib inline
import matplotlib
matplotlib.pyplot.plot(range(2,20), scores)
```

```python
clusterer5 = sklearn.cluster.KMeans(n_clusters=5)
cluster_suggestions5 = clusterer5.fit_predict(flowers['data'])
cluster_suggestions5
```

```python
clusterer5.cluster_centers_
```

```python
x = flowers['data'][:, 2]
y = flowers['data'][:, 3]
```

```python
import numpy
colouring = numpy.array(['purple', 'pink', 'blue', 'orange', 'green'])
```

```python
colour_by_cluster = colouring[cluster_suggestions5]
colour_by_cluster
```

```python
matplotlib.pyplot.scatter(x, y, color=colour_by_cluster)
```

# dbscan

```python
dbscan = sklearn.cluster.DBSCAN(eps=0.5, min_samples=3)
```

```python
dbscan_suggestions = dbscan.fit_predict(flowers['data'])
dbscan_suggestions
```

```python
import collections, numpy
```

```python
numpy.array(dbscan_suggestions)
collections.Counter(dbscan_suggestions)
```

```python
matplotlib.pyplot.scatter(x, y, color=colouring[dbscan_suggestions])
```

```python
outliers = numpy.where(dbscan_suggestions == -1, 'red', 'black')
```

```python
matplotlib.pyplot.scatter(x, y, color=outliers)
```

# Supervised learning using clusters

```python
titanic_raw = pandas.read_csv('../../SYD_DAT_7/data/titanic.csv')
```

```python
titanic = titanic_raw[titanic_raw.Age.notnull()].copy()
```

```python
titanic
```

```python
titanic['ScaledAge'] = titanic.Age / 20.0
```

```python
titanic['Gender'] = numpy.where(titanic.Sex == 'male', 1, 0)
```

```python
import sklearn.neighbors
knn = sklearn.neighbors.KNeighborsClassifier(n_jobs=5)
knn.fit(titanic[['Pclass', 'ScaledAge', 'Gender']], titanic.Survived)
```

```python
knn.predict_proba([[1, 1.0, 0]])
```

```python
knn.predict_proba([[3, 0.2, 1]])
```

```python
sacramento = pandas.read_csv('../../SYD_DAT_7/data/Sacramento-real-estate-transactions.csv')
```

```python
sacramento
```

```python
knr = sklearn.neighbors.KNeighborsRegressor(n_neighbors=5)

import sklearn.model_selection
import sklearn.cross_validation

(trainX, testX, trainY, testY) = sklearn.cross_validation.train_test_split(
    sacramento[['latitude', 'longitude', 'beds']], sacramento.price
)

knr.fit(trainX, trainY)
```

```python
predictions = knr.predict(testX)
```

```python
predictions
```

```python
matplotlib.pyplot.scatter(testY, predictions)
```

```python
matplotlib.pyplot.scatter(testY, predictions,
                            cmap=matplotlib.pyplot.cm.rainbow,
                            c=testX.beds)
```
