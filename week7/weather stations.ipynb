```python
import pandas
```

```python
weather = pandas.read_csv('../../SYD_DAT_7/data/sydney_climate.csv')
weather.dropna(inplace=True)
```

```python
weather
```

```python
# how many weather stations do we have
len(weather.columns) - 1
```

```python
# let's see if we can predict Sydney from other columns
```

```python
import sklearn.linear_model
```

```python
prediction_columns = weather.columns
prediction_columns
```

```python
prediction_columns = prediction_columns.delete(0)
```

```python
prediction_columns
```

```python
prediction_columns = [x for x in prediction_columns if x != 'SYDNEY']
prediction_columns
```

```python
X = weather[prediction_columns]
y = weather.SYDNEY
ols = sklearn.linear_model.LinearRegression()
ols.fit(X, y)
```

```python
zip(prediction_columns, ols.coef_)
```

```python
[x for x in zip(prediction_columns, ols.coef_)]
```

```python
ols = sklearn.linear_model.Lasso(alpha=10)
ols.fit(X, y)
zip(prediction_columns, ols.coef_)
[x for x in zip(prediction_columns, ols.coef_)]
```

```python
ols = sklearn.linear_model.Lasso(alpha=2)
ols.fit(X, y)
zip(prediction_columns, ols.coef_)
[x for x in zip(prediction_columns, ols.coef_)]
```

```python
elastic = sklearn.linear_model.ElasticNetCV()
elastic.fit(X, y)
```

```python
zip(prediction_columns, elastic.coef_)
[x for x in zip(prediction_columns, elastic.coef_)]
```
